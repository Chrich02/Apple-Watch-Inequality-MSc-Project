```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('deidentified_data.csv')
```

```python
plt.rcParams["figure.figsize"] = (20,3)
fig,axs = plt.subplots(1,6)
devs = data.columns[1:7]
for n in range(6):
    dev = devs[n]
    i = (data['ECG']==data['ECG']) & (data[dev]==data[dev])
    x = np.array(data.loc[i]['ECG'])
    y = np.array(data.loc[i][dev])
    ax = axs.flatten()[n]
    ax.scatter(x,y,s=0.0001)
    ax.set_title(f'{dev} (N = {len(x)})')
    ax.set_xlabel('ECG [bpm]')
    if n==0:
        ax.set_ylabel('Device [bpm]')
    ax.set_xlim((40,180))
    ax.set_ylim((40,180))
plt.show()

plt.rcParams["figure.figsize"] = (20,3)
fig,axs = plt.subplots(1,6)
devs = data.columns[1:7]
for n in range(6):
    dev = devs[n]
    i = (data['ECG']==data['ECG']) & (data[dev]==data[dev]) & (data['Skin Tone']<4)
    x = np.array(data.loc[i]['ECG'])
    y = np.array(data.loc[i][dev])
    ax = axs.flatten()[n]
    ax.scatter(x,y,s=0.0001)
    ax.set_title(f'{dev} (Lighter N={len(x)})')
    ax.set_xlabel('ECG [bpm]')
    if n==0:
        ax.set_ylabel('Device [bpm]')
    ax.set_xlim((40,180))
    ax.set_ylim((40,180))
plt.show()

plt.rcParams["figure.figsize"] = (20,3)
fig,axs = plt.subplots(1,6)
devs = data.columns[1:7]
for n in range(6):
    dev = devs[n]
    i = (data['ECG']==data['ECG']) & (data[dev]==data[dev]) & (data['Skin Tone']>3)
    x = np.array(data.loc[i]['ECG'])
    y = np.array(data.loc[i][dev])
    ax = axs.flatten()[n]
    ax.scatter(x,y,s=0.0001)
    ax.set_title(f'{dev} (Darker N={len(x)})')
    ax.set_xlabel('ECG [bpm]')
    if n==0:
        ax.set_ylabel('Device [bpm]')
    ax.set_xlim((40,180))
    ax.set_ylim((40,180))
plt.show()

```
